import the necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('house-votes-84.csv')
df.head()
```

```python
columns = df.columns
columns
```

replace ? with nan

```python
df = df.replace('?',np.NaN)
df.head()
```

```python
df.isnull().sum()
```

removing the null values

```python
from sklearn.impute import SimpleImputer

imp = SimpleImputer(missing_values=np.NaN, strategy='most_frequent')

for col in columns:
    df[col] = imp.fit_transform(df[col][:,np.newaxis])
```

```python
df.head()
```

```python
y = df.Name
X = df.drop(['Name'],axis=1)
```

changing the letters to represent some data

```python
X = X.replace('n', 0)
X = X.replace('y', 1)
X.head()
```

replacing the party names

```python
y = y.replace('republican', 0)
y = y.replace('democrat', 1)
```

```python
X.info()
```

```python
sns.pairplot(X)
```

feature selection

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectPercentile

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.75)

feat = SelectPercentile()

X_train_feat = feat.fit_transform(X_train, y_train)
X_test_feat = feat.fit_transform(X_test, y_test)
```

getting the columns for the selected features

```python
col_index = feat.get_support()
X.columns[col_index]
```

searching for the best model

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

naive = GaussianNB()
ensemb = RandomForestClassifier()
neigh = KNeighborsClassifier()

models = [naive, ensemb, neigh]

for model in models:
    model.fit(X_train_feat, y_train)
    y_pred = model.predict(X_test_feat)
    print(f'{model}: {accuracy_score(y_test, y_pred)}')
```
